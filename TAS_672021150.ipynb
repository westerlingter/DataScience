Menampilkan distribusi data umur karyawan.

yang sudah di convert ke int dan memiliki 40 jenis umur yang berbeda


```python
# Membuat plot histogram
umur = data['age']
fig = plt.figure(figsize=(10,5))
plt.hist(umur, bins=40, color='skyblue', edgecolor='black')

# Menambahkan label dan judul
plt.xlabel('Usia')
plt.ylabel('Frekuensi')
plt.title('Distribusi Usia')

# Menampilkan plot
plt.show()
```
